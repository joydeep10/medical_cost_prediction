Data pre-processing

```python
df['smoker']
```

```python
df['smoker'] = df['smoker'].map({'yes':1,'no':0})
```

```python
df.head()
```

```python
#sex and region are nominal categorical variables , so we will crate dummy variable
dummies = pd.get_dummies(df['sex'])
dummies
```

```python
#we willl use one hot encoder to feed this feature to machine
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
feature_array = ohe.fit_transform(df[['region']]).toarray()
```

```python
feature_labels = ohe.categories_
print(feature_labels)
```

```python
#creating one array
feature_labels = np.array(feature_labels).ravel()
print(feature_labels)
```

```python
#lets make dataframe of these labels
features = pd.DataFrame(feature_array, columns=feature_labels)
features.head()
```

```python
#now lets join all to df
final_df = pd.concat([df,dummies,features], axis=1)
final_df = final_df.drop(columns='region', axis=1)
final_df = final_df.drop(columns='sex', axis=1)
final_df.head()
```

now lets split the features and the target

```python
x = final_df.drop(columns='charges', axis=1)
y = final_df.charges
```

```python
y.head()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train.values, y_train.values)
```

```python
#prediction on training data
train_data_pred = reg.predict(x_train.values)
```

```python
from sklearn import metrics
r2_train = metrics.r2_score(y_train, train_data_pred)   #suitable for regression problems
print('R squared val:- ', r2_train)
```

```python
#lets predict n test data
test_data_pred = reg.predict(x_test.values)
r2_test = metrics.r2_score(y_test, test_data_pred)

print('R squared value: ',r2_test)
```

finally, now lets build a predictive system now

```python
# we select the random features in order to machine to predict respective charges
# we will also put the values assigned to categorical varibales

# features used: age:31, sex:female, bmi:25.74, children:0, smoker:no, region:southeast

input_data = (31, 25.74, 0, 0, 1, 0, 0, 0, 1, 0)
```

```python
# changing ip to numpy array and predicting the label
array_data = np.asarray(input_data)

array_data_reshaped = array_data.reshape(1,-1)

prediction = reg.predict(array_data_reshaped)
print('The insurance charge is $',prediction)
```
